```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.image as mpimg
import seaborn as sns
```

```python
#Simple examples

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Basic Matplotlib Commands

plt.plot(x,y,'r') #'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
#Creating Multiplots on Same Canvas

#plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
```

```python
#Object Oriented Method

#Create Figure (empty canvas)
fig = plt.figure()

#Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height

#Plot on that set of axes
axes.plot(x,y, 'b')
axes.set_xlabel('Set x Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')

```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #inset axes

#Large figure
axes1.plot(x,y, 'b')
axes1.set_xlabel('X_axes1')
axes1.set_ylabel('Y_axes1')
axes1.set_title('Axes 1 Title')

#Insert figure
axes2.plot(y,x,'r')
axes2.set_xlabel('X_axes2')
axes2.set_ylabel('Y_axes2')
axes2.set_title('Axes 2 Title')
```

```python
#Subplot

fig, axes = plt.subplots()

axes.plot(x,y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

```python
#Subplot with rows and cols

fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
#Interacting with axes

for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('TITLE')

fig
```

```python
#Fixing overlap

fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('TITLE')

fig
plt.tight_layout()
```

```python
#Figure size, aspect ratio and DPI

fig = plt.figure(figsize=(8 ,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x,y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

```python
#Saving figures

fig.savefig('fig.png', dpi=200)
```

```python
#Legends

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3, label="x**3")
ax.legend()
```

```python
#Plot Range

fig, axes = plt.subplots(1,3, figsize=(12,4))

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('Default axes range')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('Tight axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('Custom axes range')
```

```python
#Loading Dataset from Seaborn repository

tips = sns.load_dataset("tips")
tips.head()
```

```python
#Scater plot

ax = sns.scatterplot(x='total_bill',y='tip', data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',hue='sex',size='size')
```

```python
#Categorical Functions

sns.catplot(x='sex',y='total_bill',hue='day',data=tips,kind='strip')
```

```python
sns.catplot(x='sex',y='total_bill',hue='day',data=tips,kind='strip',jitter=False,dodge=True)
```

```python
sns.catplot(x='sex',y='total_bill',hue='day',data=tips,kind='box')
```
